### 17) There is an actor namely "El Viti". You are required to generate a report which shows all the roles played by the actor "El Viti". (Write query in SQL. Here you will have to apply JOINs and fetch the information)(Please refer the notebooks provided alongwith the class materials)

### 19) Generate a report which shows all the roles in the movie having title "P-Word Pizza" like in previous activity but this time, only fetch only those records having the role name "Pizza Guy". (Write query in SQL. Here you will have to apply JOINs and fetch the information)

```python
query = 'SELECT * FROM movie \
         JOIN role \
             ON role.movie_id=movie.id \
         WHERE \
         movie.title = "P-Word Pizza" \
         AND \
         role.name = "Pizza Guy" '
pd.read_sql_query(query, conn)
```

### 20) There is an actor namely "El Viti". You are required to generate a report which shows all the titles of movies and names of respective roles played by the actor "El Viti". (Write query in SQL. Here you will have to apply JOINs on 3 tables and fetch the information)(Please refer the notebooks provided alongwith the class materials)

```python
query = 'SELECT * FROM actor \
          JOIN movie ON (movie.id = role.movie_id ) \
          JOIN role ON (actor.id = role.actor_id ) \
         WHERE actor.name =  "El Viti" '
pd.read_sql_query(query, conn)
```
